```python
# Import the Pinecone library
from pinecone.grpc import PineconeGRPC as Pinecone
from pinecone import ServerlessSpec
import time

# Initialize a Pinecone client with your actual API key
pc = Pinecone(api_key="pcsk_3CYnJi_TZbGr8CeCcVxAsz4Li7J5n5hNBRqM7PA7k6xGKx7ftNXUYMYUJLJcb3PZrTneH4") # Replace with your actual API key


# Convert the text into numerical vectors that Pinecone can index
embeddings = pc.inference.embed(
    model="multilingual-e5-large",
    inputs=[d['question_paragraph'] for d in data],
    parameters={"input_type": "passage", "truncate": "END"})

print(embeddings)
```

```python
# Create a serverless index
index_name = "quiz-index" # Changed to lowercase

if not pc.has_index(index_name):
    pc.create_index(
        name=index_name,
        dimension=1024,
        metric="cosine",
        spec=ServerlessSpec(
            cloud='aws',
            region='us-east-1'
        )
    )

# Wait for the index to be ready
while not pc.describe_index(index_name).status['ready']:
    time.sleep(1)
```

```python
# Target the index where you'll store the vector embeddings
index = pc.Index("quiz-index")

# Prepare the records for upsert
# Each contains an 'id', the embedding 'values', and the original text as 'metadata'
records = []
for d, e in zip(data, embeddings):
    records.append({
        "id": d['question'],  # Using the question as a unique ID for simplicity
        "values": e['values'],  # The embedding values
        "metadata": {'text': d['question_paragraph']}  # Storing the question's paragraph as metadata
    })

# Upsert the records into the index
index.upsert(
    vectors=records,
    namespace="example-namespace"
)
```
